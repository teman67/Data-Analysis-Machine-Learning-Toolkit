* Understand what a NumPy **array** is

---

* NumPy stands for *Numerical Python*. It was first released in 1995 as *Numeric*, and in 2006 was relabeled as *NumPy*. 
* It is a general-purpose array-processing package, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions.

* **Why do we use NumPy?**
  * It is a **Linear Algebra** Library, and almost all of the libraries used in Data Science rely on NumPy as one of their main building blocks.
  *  It is used **to process arrays that store values**, aka data. It facilitates math operations and their vectorization. This increases performance and speeds up execution time. Handling data with a NumPy array is faster than Python lists, especially with big data sets.
  * NumPy arrays are also used to store and process **images**. TensorFlow, an ML library, use NumPy arrays for training models.

* What are the **major differences between Python lists and NumPy arrays**?
  * In a nutshell, a NumPy array is a list of lists.
  * Python lists can handle elements of different data types, whereas NumPy arrays handle only elements of the same data type.
  * Python lists consume larger memory for element calculations. NumPy arrays are comparatively more compact in memory usage and are faster than python lists. That in itself is a criterion for using NumPy, especially in big data sets. 
  * NumPy also has a big collection of mathematical and linear algebra function; which are the base of machine learning

* An array is **useful** because it helps to organize data in a grid. 
  * With this structure, elements can easily be sorted or searched.

**NumPy vs different data types**
  * An array can **handle** numbers (integer, float etc), strings (text), timestamps (dates).
  * However, how NumPy arrays handles mixed data types is different to how lists handles mixed data types

* The example below has a list with different data types, strings and integers. And when you run the cell you will see the output shows a mix of strings and integers.


my_list = ['text','label_example',55,150,'final_text_example']
my_list

* The cell below compares the data type from 2 elements from my_list

print(f"* The index 0 is:  {my_list[0]}  and its data type is {type(my_list[0])}")
print(f"* The index 2 is:  {my_list[2]}  and its data type is {type(my_list[2])}")

However this is not the case for an NumPy array, and **all the items are converted to a single data type**, we will show 2 use cases: 
 1. Not using `dtype` argument 
 1. Using the `dtype`  argument.

* First, we parsed` my_list` to` np.array()`
  * Note when you parse the list, 55 and 150 have quotes.

arr = np.array(my_list)
arr

You can check the array data type with `.dtype` method

  * Note `dtype='<U18'`, where U stands for "Unicode string" and the number next to it is the number of characters from the largest string: '`final_text_example`'
  * Unicode is a standard encoding system, used to represent characters for almost all system languages


arr.dtype

Now let's create the array with the list but this time use the `dtype` argument set to `object`
* In this specific use case, when your data is a Python list that has mixed data types, the recommendation is to use the argument `dtype`  
 * Note this time when you parse the list, 55 and 150 do not have quotes.

arr = np.array(my_list,dtype=object)
arr

You can check the array data type with `.dtype` method
  * You will notice it is 'O', or object (Python Object)

arr.dtype

And while it looks like the array contains mixed data types, this is not the case, it now contains a single data type: **dtype** of **object**
