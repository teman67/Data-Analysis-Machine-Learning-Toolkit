You can select rows or rows and columns by label using `.loc[]`
  * `.loc` stands for locate


You can select rows using its index label,
1. using a single label returns a series: 
 * df.loc['A']
1. Using a slice returns a DataFrame
  * df.loc['A': 'D']
1. Using a list of labels returns a DataFrame:
  * df.loc[['A', 'B']]

The example below grabs the row indexed/labelled as 'A'

df.loc['A']

You will notice it is a Pandas Series and has the content from all columns

### iloc 

Alternatively, you may be interested to subset your data frame based on index location, not on location.
You can use .iloc[], which means index location.
It accepts the indices of the desired rows and columns. The pandas dataframes have integer index for both rows and columns

To select you can:
1. Use a single integer for a single row:
 * df.iloc[0]
1. Use two integers for a value in a row column combination:
 * df.iloc[0, 2]
1. Use a list of integers for rows only:
 * df.iloc[[0,1,2]]
1. Use lists of intergers for rows and columns:
 * df.iloc[[0,1,2], [2,3]]
1. Use a slice object on rows:
 * df.iloc[:3]
1. Use a slice object on rows and columns:
 * df.iloc[:2, 3:]



The example below grabs the row at index 0. 
  * In this case, you don't have to know/remember the index label from the row at index 0 is "A"

df.iloc[0]

df

You can subset ranges, using notation similar to NumPy
* The example below gets rows until index 2 (not including index 3) and columns until index 2 (not including index 3)
  * Remember that "A" is row index 0 and "Col1" is column index 0

df.iloc[:3,:4]

### .query()
Query the columns of a DataFrame with a **boolean expression**.

* Quick recap on `df` first.

df

You can select data in your DataFrame using notation similar to NumPy
* The example below creates a boolean series on `Col4 > 0 `and is parsed at df

 * You will notice that row "C" is not selected, since the value in row C at Col4 was negative
 * Think of it like: I want the values from Col4 that are greater than 0 

df[df['Col4']>0]

There is an alternative pythonic way to subset data using `.query()`
  * You parse as a string the operation you are interested in

df.query("Col4 > 0")

You can combine multiple operations.
  * The example below considers data where `Col4 > 0 and Col1 > 0.5`
  * If you want to use an or operation, just write `or`

df.query("Col4 > 0 and Col1 > 0.5 ")

Let's recap `df`

df

* Note the difference between the outputs below

First case

df.query('Col2 > 0.8 or Col2 < 0 and Col1 > 0.5')

Second case

df.query('(Col2 > 0.8 or Col2 < 0) and Col1 > 0.5')

Why did that happen?
  * The parentheses in combination to the operators (`or` and `and`) gave the priority to select the orders to execute the query command
    * The first case, first it queried `Col2 < 0 and Col1 > 0.5`, then `Col2 > 0.8`
    * The second case, first it queried `Col2 > 0.8 or Col2 < 0`, then the result was queried with `and Col1 > 0.5`
