### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Index
Let's consider the DataFrame below

np.random.seed(101)
df = pd.DataFrame(data=np.random.randn(5,4),
                  index=['A', 'B', 'C', 'D', 'E'],  
                  columns=['Col1', 'Col2', 'Col3', 'Col4']
                  )
df

You can check index content with the method `.index`



df.index

You can reset your index label to integers starting from 0 using `.reset_index()`
  * This example doesn't overwrite the DataFrame, since `inplace` is not set to `True`


df.reset_index()

You will notice that the `index` would become an additional column in the DataFrame.
  * If you are not interested in keeping the index column, you can set `drop=True`
  * The example above doesn't overwrite the DataFrame. 

The example below overwrites the DataFrame since it includes `inplace=True` (a parameter you are familiar with already)

  * You can check the DataFrame content and note that the labels A, B, C, D, E are not indices. 
  * It is now a range of integers starting from 0. In addition, the labels A, B, C, D, E no longer exist

df.reset_index(drop=True, inplace=True)
df

df['Col5'] = ['Part1','Part2','Part3', 'Item1','Item2']
df

Let's set Col5 as index with `.set_index()` and `inplace=True`

df.set_index('Col5',inplace=True)

Your DataFrame can have **MultiIndex** 
  * First, consider the following DataFrame . 

np.random.seed(101)
df = pd.DataFrame(data=np.random.randn(5,4),
                  index=['A', 'B', 'C', 'D', 'E'],  
                  columns=['Col1', 'Col2', 'Col3', 'Col4']
                  )
df

We will check the index with `df.index`.  
  * You will notice there is only one level

df.index

---

Now consider the DataFrame below made with NumPy functions you are already familiar with.

np.random.seed(seed=13)
df = pd.DataFrame(data={'Col1':np.random.randint(low=10, high=20, size=10),
                        'Col2':np.random.randint(low=-10, high=0, size=10),
                        'Category1': np.random.choice(['X','Y','Z'], size=10),
                        'Category2': np.random.choice(['a','b','c','d','e','f','g'], size=10)}
                  )

df

We will set Category1 and Category2 as Index with `.set_index()`.
  * Let's see the effect of it.
    * Note there are 2 levels of the index now when you hover the mouse on the index area

df.set_index(['Category1','Category2'], inplace=True)
df

Now we check index content and see the mention of `MultiIndex`, and the respective pairs of Category1 and Category2 that create together this `MultiIndex` structure

df.index

We can grab the index names with `.index.names`

df.index.names

The MultIindex format will likely appear over your career in the workplace. 
  * Typically, **we tend to prefer** a DataFrame that is not MultiIndexed, since the majority of the data science packages have functions/methods more suitable for non MultiIndex DataFrames.
  * In this case, you can remove the hierarchical index by adding `.reset_index()`

df.reset_index(inplace=True)
df
