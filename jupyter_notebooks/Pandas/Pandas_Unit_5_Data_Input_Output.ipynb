### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> CSV files

A CSV file, or comma separated values, is simply a text file that allows the data to be saved in a table structured format and uses a comma to separate values.

image.png

You can read and save data in multiple formats using Pandas.
* You can use `pd.read_csv()` function to read csv files. 
  * There are multiple arguments in this function, you can check the list of them [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html). For this case, we will use the first argument `filepath_or_buffer` and provide the path where the csv file is.
  * Generally, for this function, we don't explicitly write `filepath_or_buffer` argument, since over time you will get used to this function and will associate the first argument is where you write the file path

You can get tables from URLs and process like DataFrames


We are interested to read the table from the [FIFA World Cup page on Wikipedia](https://en.wikipedia.org/wiki/FIFA_World_Cup) using `pd.read_html()`. You should parse the link.
  * The central aspect in the URL, is to contain tables, so the function can read the tables and convert to DataFrames
  * Please also visit the link above for few minutes, so you can get familiar with the content

search = pd.read_html('https://en.wikipedia.org/wiki/FIFA_World_Cup')

Let's investigate the "`search`" content. You will notice it is not a DataFrame

search

You will notice the type of the answer is not a DataFrame, but a Python list

type(search)

And with `len()` you can discover how many items this list has

len(search)

Each item in this list is a DataFrame. The function `pd.read_html()` gets all tables in that link, transform to DataFrame and append in Python lists
* It will be your job to find the DataFrame you are interested in by searching across the list elements
* Let's investigate the first list element

search[0]

Let's try the last item

search[24]

For this exercise, we will be happy with `search[3]`, showing the soccer world cup editions, indicating the year, hosts and top 4 nations
* You will notice the dataframe as it needs to be processed to extract meaningful insights. This exercise's objective is to read the table from an URL, not to process the data.
  * For example, the first row is, in fact, the columns name. 
  * The columns with only NaN indicates blank spaces into the original HTML table.

df = search[3]
df
