Consider the DataFrame below, created with a Python dictionary, with 4 columns and content generated with NumPy array (numerical data) and Python List (categorical data)
* Col4 is a set of random boolean (`True,False`) generated with NumPy
* The index is made with `pd.date_range()` function, where you create a series of datetime data. This function is useful when you want to create datetime data and its documentation is [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html). In this example, we use the "`start`" argument as 01/01/2021, set 10 "periods" and defined period frequency with "`freq`" equals to "M". 
  * "M" stands for month-end frequency, so the days in the dates will be the last day in that given month. For example, you will notice that for January, it will be 31, and for February, 28. 
* The "`freq`" values are also called offset-aliases and can receive different values depending on your interest, like weekly frequency, hourly frequency, month-end frequency etc. You can find the possible choices [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases).

df = pd.DataFrame(data={'Col1':np.random.randint(low=0,high=50,size=10),
                        'Col2':np.random.randn(10),
                        'Col3':['A','B','C','A','A','C','D','B','A','A'],
                        'Col4':np.random.choice([True, False],size=10),
                        'Col5':np.random.randint(low=1,high=4,size=10)},
                  index=pd.date_range(start='01/01/2021', periods=10, freq='M'))
df

## <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> .head() and .tail()

Rather than displaying a big dataset just to see the initial data rows you can instead inspect the DataFrame using `.head()` by returning the first n rows 
  * By default, it displays the first 5. The function documentation can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html)

df.head()

At the same time you can set with "`n`" argument. This example you want to see the first 3

df.head(n=3)

Similarly, if you want to see the most recent data you can return the last n rows with `.tail()`. 
  * By default, it displays the last and like the previous example, you can set `n`. The function documentation can be found [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.tail.html)

df.tail()

## <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> .corr()

You can assess correlation levels across **numerical** variables with `.corr()`. By default, the method is '`pearson`', which is a measure of the linear relationship between two features
  * The function documentation can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html)

df.corr()  # or df.corr(method='pearson')

You can set `method='spearman'` to assess monotonic relationships (whether linear or not) in numerical variables. Monotonic is where one variable is associated with a change in the specific direction of another variable. e.g. Does a positive change in value X result in a positive change in value Y.
* There will be a dedicated lesson on how to interpret correlation levels

df.corr(method='spearman')
