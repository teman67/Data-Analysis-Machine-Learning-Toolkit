You can reshape the array without changing the data within it using  `.reshape()` method.
  * The example below shows a 1-d array, with 30 elements. 

np.random.seed(seed=0)
arr = np.random.randint(low=1,high=150,size=30)
arr

You can reshape as 3 rows x 10 columns, transforming it into a 2-d array
  * The argument is the shape you are interested in

arr_reshaped = arr.reshape(3,10)  # 2-d array, 3 x 10
arr_reshaped

* When reshaping, the size should fit into the new shape. For example, considering the previous array, you can't reshape it to 3 x 15

There can be a situation where you have a multidimensional array and want to transform it into a 1-d array. You can use `.flatten()` to do this.
* Consider a 2-d array made with `np.random.randint()`

np.random.seed(seed=0)
arr = np.random.randint(low=1,high=150,size=(2,5))
arr

You can flatten it, and you will notice it becomes a 1-d array. 
* No arguments are needed for the `.flatten()` method

arr_flatten = arr.flatten()
arr_flatten

Alternatively, you can get the same effect using `.reshape()`, with the argument as -1
* Setting -1 at `.reshape()` indicates you want to reshape to a 1-d array but you dont specify the length, you let the computer calculate by itself

arr_reshape = arr.reshape(-1)
arr_reshape

 You can check the shape and type of a NumPy array using, respectively, the attributes `.shape` and `.dtype` 
* Consider a 2-d array (5x2) made using `np.arange()`

arr = np.arange(start=1,stop=11,step=1).reshape(5,2)
arr

We combined a `print()` statement, f-string, `.shape` and `.dtype` to demonstrate the use case to check an array's shape and dtype


print(
    f"* Array:\n {arr} \n\n"
    f"* Array shape: \n {arr.shape} \n\n"
    f"* Array type: \n {arr.dtype}"
    )

Consider the array below, not made with integers

np.random.seed(seed=0)
arr = np.random.randn(2,4)
