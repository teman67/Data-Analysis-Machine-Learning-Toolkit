### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Txt file

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> A text file is a file that is structured as a sequence of lines. Typically it contains little or no formatting. Its extension is `.txt`

Let's imagine you have locally a txt file that you need to analyze.
* Python has a built-in function called `open()`. The documentation is [here](https://docs.python.org/3/library/functions.html#open). In the exercise, we will use the argument:
  * file: for the file path (in our case, the file is on the root)
  * mode: `r` stands open for reading. Check the documentation to see other options

We will use the `with` structure to open the file; then we will read the file using `.read()` and store it in a `txt` variable. Once the code exits the `with` structure, it closes the file

with open(file='file_1.txt', mode='r',  encoding='utf8') as f:
    txt = f.read()

txt

Let's check the files directory again

! ls

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> We can now save the content from all txt files to a python list to analyse them later.
* We use a Python library called glob to show the current files in the directory and specify the type of files we want to consider; in this case, all txt files (`*.txt`). The glob documentation is [here](https://docs.python.org/3/library/glob.html)

import glob
glob.glob(pathname="*.txt")

We create the logic where we loop over the files, and for each, we append the content to a list (`list_txt_files`)

list_txt_files = []

for txt_file in glob.glob(pathname="*.txt"):
  with open(file=txt_file, mode='r') as f:
    lines = f.read()
    list_txt_files.append(lines)

list_txt_files

### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> SQL

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> SQL stands for Structured Query Language and is  used to manage data in a relational database management system (RDBMS).
* It is particularly useful in handling structured data, for example, data incorporating relations among entities and variables.

You will likely need to pull data from SQL databases in the workplace; however, it will not be required for this course. 
You can use many Python libraries to pull data from SQL, like pyodbc or PyMySQL. As a result, the following section is text-based, meaning you will not see runnable examples. We will display only the code for pulling data from SQL using PyMySQL, so you will know which library to consider when you face a similar task in the workplace. PyMySQL has a handy API interface. The documentation link for the library is [here](https://pymysql.readthedocs.io/en/latest/). We will also show code (not a runnable example) on how to use Pandas to pull the data from the database

* We need to connect the session to a database before pulling the data; we shall provide:
  * server name
  * database name
  * user
  * password


import pymysql.cursors  
server = ""  
database_name = ""  
user = ""   
password = "" 

#### Connect to the database
We create a connection using pymysql.connect(). We should provide the server, database_name, user and password.  

<pre>
connection = pymysql.connect(host=server,
                             user=user,
                             password=password,
                             database=database_name,
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor
</pre>

##### Delete your SQL database credentials from memory
del server, database_name, user, password 

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> You can then perform a SQL query you are familiar with, for example: select, insert, update, and delete. 
* Then we parse the query together with the connection object (connection) at `pd.read_sql()` and store the result in a variable.
* Next, we close the connection. This is a practice you should follow when exchanging data with a SQL database; always remember to close the connection. Otherwise, that will impact the database performance and may jeopardize other users that are using the database
* The pulled data is now stored in a DataFrame named “result”

<pre>
import pandas as pd

sql_query = f"SELECT * FROM your_table"

result = pd.read_sql(sql_query, connection)
connection.close()
</pre>
