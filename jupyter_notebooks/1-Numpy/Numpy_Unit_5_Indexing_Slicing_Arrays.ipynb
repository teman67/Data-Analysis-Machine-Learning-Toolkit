````python
import numpy as np

# Create a simple 1-d array using np.arange()
arr = np.arange(start=0, stop=26, step=2)
arr

# Retrieve the first element
arr[0]

# Retrieve a range of elements from index 5 to 10 (exclusive)
arr[5:10]

# Retrieve elements from index 7 until the end
arr[7:]

# Create a 2-d array using np.random.randint()
np.random.seed(seed=101)
arr = np.random.randint(low=10, high=50, size=(3, 5))
arr

# Retrieve a full column (considering all rows)
arr[:, 3]

# Grab a range of elements
arr[1:, 3:]

# Create a 3-d array using np.random.randint()
np.random.seed(seed=101)
arr = np.random.randint(low=10, high=50, size=(3, 3, 4))
arr

# Retrieve a column
arr[0, :, 2]

# Retrieve a range of elements
arr[0, :2, :3]

# Grab more than one array
arr[1:3, :, 1:3]

# Retrieve a specific range of elements
arr[1:3, :, 1:3]
````
