As an alternative for slicing, **negative slicing** retrieves elements **starting counting, from the end**, rather than the beginning.
* Consider a 1-d array made with `np.random.randint()`

np.random.seed(seed=101)
arr = np.random.randint(low=10,high=50,size=9)
arr

Retrieve ***last 3 elements***
  * "-3:" means: count 3 index position (inclusive), from the end. You will start moving right and it goes until the end since there is no number after ":"

arr[-3:]

Retrieve the following ***elements: 19 and 23***
  * "-3:-1" means: count 3 index position (inclusive), from the end. You will start moving right and it stops 1 before the end

arr[-3:-1]

---

Consider a 2-d array made with np.random.randint()

np.random.seed(seed=101)
arr = np.random.randint(low=10,high=50,size=(3,5))
arr

Retrieve ***all rows and the last 2 columns***
* Slicling technique:
    * Rows: You will get all rows, due to the first ":"
    * Columns: "-2:" means: count 2 index position (inclusive), from the end. You will start moving right and it goes until the end since there is no number after ":"

arr[:,-2:]

````python
import numpy as np

# 1-d array made with np.random.randint()
np.random.seed(seed=101)
arr = np.random.randint(low=10, high=50, size=9)
arr

# Retrieve last 3 elements
arr[-3:]

# Retrieve elements 19 and 23
arr[-3:-1]

# 2-d array made with np.random.randint()
np.random.seed(seed=101)
arr = np.random.randint(low=10, high=50, size=(3, 5))
arr

# Retrieve last 2 rows and 3rd and 4th columns
arr[-2:,-3:-1]

# 3-d array made with np.random.randint()
np.random.seed(seed=101)
arr = np.random.randint(low=10, high=50, size=(3, 5, 4))
arr

# Retrieve the second array, its 3rd and 4th rows, and last 3 columns
arr[-2, -3:-1, -3:]
````
