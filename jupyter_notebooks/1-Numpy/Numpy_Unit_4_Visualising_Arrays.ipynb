Translating the array dimensions to a more playful and visual manner will help to **make sense** of NumPy arrays.

image.png

For example, a **1-d array** can be visualized as a **length**. If your 1-d array has 8 elements, imagine these elements arranged in a line.



np.random.seed(seed=1)
np.random.randint(low=10,high=50,size=(6))

---

image.png

A **2-d array** can be visualized as an arrangement of **rows and columns**. 
* If your 2-d array is 3 by 2, you can visualize it as 3 rows and 2 columns


np.random.seed(seed=1)
np.random.randint(low=10,high=50,size=(3,2))

---

image.png

The Jupyter Notebook cell's output may not sometimes facilitate to visualise a **3-d array**
* A **3-d array** can be visualized as **a set of arrays, "arranged" on top of each other**. 
  * You will read from ***left to right, like number of arrays + 2-d array***.
  * For example, if your 3-d array is 2 by 3 by 4, that means you can visualize it as 2 arrays of 3 by 4. 
  * If your 3-d array has a shape of (3,4,3), imagine 3 arrays of 4 by 3


np.arange(start=0,stop=36).reshape(3,4,3)

---

Can I have a 4-d array or more?
* You may have as many dimensions as you would like to.
* The example below shows a 4-d array:

np.random.seed(seed=1)
np.random.randint(low=10,high=50,size=(2,3,4,5))
