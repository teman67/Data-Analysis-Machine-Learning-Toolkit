You can add more conditions to your selection. Imagine if you want values greater than 4 and less than 17 (but including 17)

Combining Two Selection 
* Imagine if you want values smaller than 2 (including 2) or greater than 18. You also want values that are equals to 9
  * This example combines 2 selections with an **or** operator **"|"**
    * ( (arr<=2) | (arr > 18) )
    * (arr==9)

arr[ ((arr<=2)|(arr > 18)) | (arr == 9) ]

---

````python
import numpy as np

# 2-d array
np.random.seed(seed=10)
arr = np.random.randint(low=1, high=25, size=(3,5))
arr

# Array elements that are less than 7
arr[arr < 7]

# Array elements that are less than 7, reshaped to 1-d array
arr[arr < 7].reshape(-1)

# Array elements that are greater than 4 and less than or equal to 17
arr[(arr > 4) & (arr <= 17)]

# Array elements that are smaller than or equal to 5 (including 5) or greater than 15
arr[(arr <= 5) | (arr > 15)]

# Array elements that are smaller than 2 (including 2) or greater than 18, or equal to 9
arr[((arr <= 2) | (arr > 18)) | (arr == 9)]
````
