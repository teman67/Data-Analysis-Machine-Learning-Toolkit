**Lambda functions**

* There are use cases where the simple way is enough, for instance, if we have a simple sum logic to create a column
* But there are circumstance where the logic is more sophisticated



Consider the simple function below that adds 10 to your value

def my_function(x):
  return x + 10

variable = my_function(45)
variable

You can do that in one line with `lambda` function, or also know as lambda expression

variable = (lambda x: x + 10)(45) 
variable


The Lambda function is a small function that can take any number of arguments but can only have one expression.
* The structure of a lambda function is:
  * **lambda arguments : expression**
  * you are not limited to one argument/variable when using a lambda, you can add more, in this case, we have x and y
    * In the example below, the function is x + 3*y, and you parsed x=2 and y=1

variable = (lambda x,y: x + 3 * y)(2,1)
variable

I have the logic to define the value of a diamond based on the cut and price. Say this logic is due to my experience with diamonds
  * I create a function called `my_price_perception()`


def my_price_perception(price,cut):
  if cut in ['Premium','Ideal'] and price < 2000:
    return "Worth it, buy now"
  elif price < 1000:
    return 'Buy'
  elif price < 2500:
    return 'Think twice'
  else:
    return 'No thanks'


We create a column '`MyValuePerception`' by using an apply function with lambda along the row with `axis=1`
  * the argument x for lambda function is df, our DataFrame
  * for `my_price_perception` function, we parse the price and the cut.

df['MyValuePerception'] = df.apply(lambda x: my_price_perception(x['price'],x['cut']),axis=1)
df

We can check now how many items you should buy! :)
