You can rename your columns using 2 approaches
  * assigning to df.columns a list with the column names you want
  * or using `.rename()`


First we create a separate DataFrame for this example

df_example = df.copy()
df_example.head()

Then we assign to `df.columns` a list with the column names we want

df_example.columns = ['Revenue', 'Rate', 'ProfileCategory', 'Flag', 'ClassCategory']
df_example.head()

Alternatively you can use `.rename()`. The documentation is [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html). You should parse a list mapping the old column names and new column names at 'columns' argument. Note we set `inplace=True`, to overwrite the DataFrame



  * This is **useful and can save time** when you have a long list of columns and don't want to list all column names. You just map to rename the columns in which you are interested in changing. In this example we just mapped 3 columns. The others remain the same.

df_example = df.copy()
df_example = df_example.rename(columns={"Col1":"Revenue",
                                        "Col2":"Rate",
                                        'Col4':'Flag'})
df_example.head()

We can use the argument `inplace=True` as an alternative. This creates a new DataFrame. The difference is that you dont have to reassign the variable.

df_example = df.copy()
df_example.rename(columns={"Col1":"Revenue",
                           "Col2":"Rate",
                           'Col4':'Flag'},inplace=True)
df_example.head()

You can select data based on data type with `.select_dtypes()`. This function documentation is [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.select_dtypes.html)
* Say you are interest to subset only numerical data, you should set "`include`" argument as '`number`'
* You might do this if you want to create a scatterplot 

df.select_dtypes(include='number')

Now you are interest to subset only categorical data, you should set "`include`" argument as '`object`'

df.select_dtypes(include='object')

If you want boolean data, set '`boolean`' for '`include`'

df.select_dtypes(include='boolean')

If you want a combination of multiple data types, parse them in a list.
  * The example below considers boolean and number types

df.select_dtypes(include=['boolean','number'])

For the next example we create a separate DataFrame df1, and reset the index of df.
  * The previous index became a column called '`index`'

df1 = df.reset_index()
df1

We want now to select, from df1, the columns where datatype equals to '`datetime`'

df1.select_dtypes(include='datetime')

You can also a reverse logic, use `exclude` parameter when you don't want a specific data type in your DataFrame.
* Naturally, you can use it for all data types we showed earlier
* We will showcase with 'number' only

df.select_dtypes(exclude='number')
