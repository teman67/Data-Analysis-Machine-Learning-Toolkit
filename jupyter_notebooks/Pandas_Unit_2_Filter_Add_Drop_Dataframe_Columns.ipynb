### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Filter with .filter()

Another way to subset data, in a more pythonic way, in a  DataFrame is with `.filter()`. Just parse in a list with the columns you are interested in.
* The example below retrieves Col1

df.filter(['Col1'])

The example below retrieves 2 columns: Col1 and Col3
  * The columns are parsed in a list
  * The argument is `axis=1`, indicating you are grabbing columns

df.filter(['Col1','Col3'],axis=1)

You can drop columns or rows from your DataFrame, using `.drop()`

  * You set `labels` as your columns or index. `Axis=1` indicates you want to drop columns, `axis=0`, rows.
* The example below drops Col5

df.drop(labels=['Col5'],axis=1)

Check again `df` content

df

You will notice Col5 is still there. 
  * There are 2 ways to solve: using `inplace=True` or adjusting the code

* If you add `inplace=True` to  `df.drop(labels=['Col5'],axis=1)`, then you are **overwriting the existing DataFrame**


df.drop(labels=['Col5'],axis=1,inplace=True)
df

Second approach: Or you can write `df = df.drop(labels=['Col5'],axis=1)`, in this way you reassigin the content of df.
  * But first, let's re-add Col5 to conduct the exercise

df['Col5'] = df['Col1'] + df['Col3']
df

You can drop rows using index reference, setting `axis=0`
  * The use case is similar, at `labels` you parse in a list the indices you want to drop

df.drop(labels=['A','E'],axis=0)

Remember the example above doesn't remove "for real" the rows since the `inplace` is not set as `True` or the DataFrame was not reassigned
