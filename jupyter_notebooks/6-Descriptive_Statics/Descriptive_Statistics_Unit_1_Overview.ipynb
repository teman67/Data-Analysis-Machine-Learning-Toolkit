#### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Numerical

image.png

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> Numerical variables where values are numbers, and they can be discrete or continuous. The difference between these relies on the following:
* A continuous variable may contain any value in a range. For example, how much you spent in the market this week, the price of your mobile phone or the time, in seconds, you spent today in this lesson. These values can be a fraction, so that you may have spent 52.34 euros this week in the market.
* Discrete means that there are only particular valid values, for example:
 - number of times you went to the market this week (in this case the values are not fractions)
 - shoe size, like 6, 6.5, 7, 7.5, 8, 8.5 (note the values are fractions, but the values are not in a continuous range, only particular values are allowed)


---

#### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Categorical

image.png

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> Categorical variables are selected from a group of labels. For example, gender can be male or female. Marital status can be single, married, divorced or other. Colour can be yellow, red, purple or blue.
* The cardinality is the number of different labels it can have. In our previous examples, gender was 2, and marital status 3 and colour 4


<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> They can be broken down into **ordinal** or **nominal**.
* An ordinal variable can be related to an order, like days of the week or student grade (fail, pass, pass with honour). 
* On the other hand, nominal variables dont relate to a specific order, like preferred colour: which can be either blue, green, yellow, and in this case, there is no specific order or rank.



<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> In your career, you might be in a situation where a categorical variable will be encoded as a number; for example, gender can be encoded as 0 for female and 1 for male. 
* The data practitioner should consider the context and realize that in this case, it is not a discrete variable but a categorical variable that was encoded
* Other edge case is for IDs. Typically when you subscribe for a service, say a mobile phone, the company may generate a unique ID for you. This ID may be a set of numbers, but that doesnt mean it makes sense to perform calculations with it. It may be better framed as a categorical variable

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png">  Descriptive statistics is about describing and summarizing the data:
  * via a **quantitative approach** (numerical summary) and 
  * using a **graphical representation** (plots).



<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> We are interested to **summarize the dataset's variables**. 
  * If you describe one variable, that is univariate analysis.
  * If you study two variables at once, bivariate. 
  * If you study more than 2, multivariate.



<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> We typically break down descriptive statistics into 2 major studies:
  * **Central tendency**
  * **Variability**

### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Population and Sample

image.png

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> When analyzing data, we should be aware of the **population and sample** concept.
  * Population is all existing elements from that particular variable. In practical terms, a population has many elements, making it hard to collect all. 
    * Imagine if you want to collect people's preference on pizza, hot dog or barbecue in your country. That will likely be a lot of people. Collecting data from a lot of people means a lot of energy, time and money.
  * On the other hand, a sample is a part of a population, which ideally preserves the statistical characteristics if the sample size is statistically significant. That allows us to draw inferences about the population based on relevant samples. The larger the sample, the smaller the error when compared to the whole population.



### <img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%2010-%20Lesson%20Content.png"> Outliers

image.png

<img width="3%" height="3%" align="top"  src="https://codeinstitute.s3.amazonaws.com/predictive_analytics/jupyter_notebook_icons/Icon%207-%20Note.png"> In the workplace, you will be faced with outliers, which are simply data points different from the majority of the data. 

* To understand outliers, we would have to come back to the data collection moment and make sense of its context and conditions, since maybe an error might have happened for that given data point, or eventually, that data point is an observation of new behaviour. However, sometimes it is difficult to access the data collection moment,

* There may be multiple reasons for outliers, like a new natural behaviour of the data, an error in the data collection process, due to eventually human error/bias, equipment not calibrated etc

* There is a separate section in this lesson to make sense of outliers. A future lesson will dive further into how to handle outliers; the scope now is to make sense of it.
